### Single Time Series Plot

```python
### Load Data
data_to_plot = pd.read_csv(
    "./data/13F-cover-pages.csv", parse_dates=["periodOfReport", "filedAt"]
)


### Helper Functions
def get_quarter(period):
    if period.month == 3:
        return "Q1"
    elif period.month == 6:
        return "Q2"
    elif period.month == 9:
        return "Q3"
    elif period.month == 12:
        return "Q4"
    else:
        return "Q0"


def format_period(period):
    qtr = get_quarter(period)
    if qtr != "Q1":
        return qtr
    return str(period)[0:4] + " " + get_quarter(period)


### Plotting Function
def plot_holding_values(data):
    fig, ax = plt.subplots(figsize=(5.5, 3))

    data_grouped = data.groupby("periodOfReport")
    data_grouped = (
        data_grouped["holdingsValueStandardized"].sum().to_frame().reset_index()
    )

    ax.plot(
        data_grouped["periodOfReport"],
        data_grouped["holdingsValueStandardized"],
        color="black",
    )

    xticks = data_grouped["periodOfReport"].dt.date.unique().tolist()
    ax.set_xticks(xticks)
    ax.set_xticklabels(xticks, rotation=90)
    ax.yaxis.grid(True, linestyle="--", which="major", color="grey", alpha=0.25)
    ax.spines["top"].set_visible(False)
    ax.spines["right"].set_visible(False)

    # draw vertical lines for every Q1. e.g. 2013-03-31, 2014-03-31, 2015-03-31, ...
    y_min = ax.get_ylim()[0]

    for year in data_grouped["periodOfReport"].dt.year.unique():
        x_value = str(year) + "-03-31"
        x_value = pd.to_datetime(x_value, format="%Y-%m-%d")
        if (x_value in data_grouped["periodOfReport"].values) == False:
            continue
        y_value = data_grouped[data_grouped["periodOfReport"] == x_value][
            "holdingsValueStandardized"
        ].values[0]

        ax.vlines(
            x=x_value,
            ymin=y_min,
            ymax=y_value,
            color="grey",
            alpha=0.25,
            zorder=-100,
            linestyle="--",
            linewidth=0.8,
        )

    ax.set_ylabel("Value of all Holdings\n(USD, Billions)")

    ax.get_yaxis().set_major_formatter(
        mtick.FuncFormatter(lambda x, p: format(int(x) / 1000000000, ",.0f"))
    )

    ax.axvspan(
        pd.to_datetime("2020-03-31", format="%Y-%m-%d"),
        pd.to_datetime("2021-12-31", format="%Y-%m-%d"),
        alpha=0.1,
        color="red",
        zorder=-100,
    )

    ax.text(
        pd.to_datetime("2020-12-31", format="%Y-%m-%d"),
        ax.get_ylim()[1] * 0.95,
        "COVID",
        horizontalalignment="center",
        verticalalignment="center",
        color="red",
        alpha=0.5,
    )

    total_number_of_funds = data["filerCik"].nunique()
    number_of_13F_filings = data["accessionNo"].nunique()

    caption = (
        "Figure: Total Value of all Holdings per Quarter. The figure shows the total dollar value of all holdings per quarterly reporting period (end of March, June, September and December). "
        + "The value is calculated by summing the value of all holdings reported in a given quarter per investment manager. The value is standardized to USD billions. "
        + f"A total of {format(total_number_of_funds, ',')} investment managers and {format(number_of_13F_filings, ',')} 13F filings are included in the figure. "
    )

    caption = "\n".join(textwrap.wrap(caption, 70))
    fig.text(0.5, -0.35, caption, ha="center", fontsize=9)

    ax.set_xticklabels(
        [format_period(period) for period in xticks],
        rotation=90,
        ha="center",
        fontsize=8,
    )

    plt.tight_layout()
    file_path = "./assets/all-holdings-values-per-quarter.png.pdf"
    plt.savefig(file_path, dpi=300, format="pdf")
    print(f"Saved to {file_path}")
    plt.show()


plot_holding_values(data_to_plot)
```

### Histogram

```python
### Load Data
funds_filings = pd.read_csv('./data/13F-filings-metadata.csv')

ciks = funds_filings["cik"].unique()

years_active = []

for cik in ciks:
    years = funds_filings[funds_filings["cik"] == cik]["year"].unique()
    years_active.append({"cik": cik, "years_active": len(years), "years": years})

years_active = pd.DataFrame(years_active)

fig, ax = plt.subplots(figsize=(4, 3))

bins_range = range(1, 27)
ax.hist(years_active["years_active"], bins=bins_range, color="black")

# move all bars by 0.5 to the left
ax.set_xticks(np.array(bins_range) - 0.5)
ax.set_xticklabels(bins_range, rotation=90, fontsize=8, ha="center")

ax.yaxis.grid(True, linestyle="--", which="major", color="grey", alpha=0.25)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_xlabel("Years Active")
ax.set_ylabel("Number of Funds")

caption = (
    "Figure: Distribution of Life Span of Funds (1998 - 2023). The figure shows the distribution of "
    + "the number of years a fund was active. A fund is considered active if it filed a 13F-HR form "
    + "in a given year. Funds are counted based on their CIK."
)

caption = "\n".join(textwrap.wrap(caption, 60))
fig.text(0.5, -0.30, caption, ha="center", fontsize=9)

plt.tight_layout()
file_path = "./assets/fund-life-span-distribution.png.pdf"
plt.savefig(file_path, dpi=300, format="pdf")
print(f"Figure saved to {file_path}")

plt.show()

print("Descriptive Statistics of the Life Spans of Funds (1998 - 2023)")
pd.DataFrame(years_active["years_active"].describe().round(2))
```

### Multiple Plots in a Grid

```python
### Load Data
alpha_original_vs_cloned = pd.read_csv('./data/13F-alpha-original-vs-cloned.csv')


fig, ax = plt.subplots(2, 1, figsize=(6, 4), sharex=True)

for i, a in enumerate(ax):
    x = (
        alpha_original_vs_cloned["original"]
        if i == 0
        else alpha_original_vs_cloned["cloned"]
    )
    a.set_title("Original Funds" if i == 0 else "Cloned Funds")
    stats_x = pd.DataFrame(x.describe().round(2))
    stats_x =x.describe().round(2)
    # calculate skewness and kurtosis
    skewness = x.skew()
    kurtosis = x.kurtosis()
    # print onto figure the mean, median, and standard deviation of alpha
    stats_text = (
        f"Funds: {stats_x['count']:,.0f}\n"
        + f"Mean: {stats_x['mean']}%\n"
        + f"Std. Dev.: {stats_x['std']}%\n"
        + f"25%: {stats_x['25%']}%\n"
        + f"50%: {stats_x['50%']}%\n"
        + f"75%: {stats_x['75%']}%\n"
        # skewness and kurtosis
        + f"Skewness: {skewness:.2f}\n"
        + f"Kurtosis: {kurtosis:.2f}"
    )
    a.text(
        0.95,
        0.99,
        stats_text,
        verticalalignment="top",
        horizontalalignment="right",
        transform=a.transAxes,
        fontsize=8,
    )
    a.hist(x, bins=100, color="black", alpha=0.9)
    a.set_yscale("log")
    a.set_ylabel("Number of Funds")
    a.axhline(y=10, color="grey", alpha=0.5, linestyle="--", linewidth=0.8, zorder=-100)
    a.axhline(
        y=100, color="grey", alpha=0.5, linestyle="--", linewidth=0.8, zorder=-100
    )
    a.axvline(x=0, color="red", alpha=0.5, linestyle="--", linewidth=1.8, zorder=100)
    a.grid(False)


ax[1].set_xlabel("Alpha (%)")
# set title of figure "Alpha Distributions of Original and Cloned Portfolios (2013 - 2023)"
# fig.suptitle("Alpha Distributions of Original and Cloned Portfolios (2013 - 2023)")


plt.tight_layout()
file_path = "./assets/alpha-distributions.png.pdf"
plt.savefig(file_path, dpi=300, format="pdf")
print(f"Saved to {file_path}")

plt.show()
```

### Multi Scatterplots

```python
from scipy.stats import wilcoxon, linregress

### Load Data
data_to_plot_original = pd.read_csv("./data/13F-factors-original.csv")
data_to_plot_cloned = pd.read_csv("./data/13F-factors-cloned.csv")

columns_of_interest = [
    "alpha",
    "beta",
    "ariMeanRt",
    "geoMeanRt",
    "cagr",
    "sharpe",
    "sortino",
    "annVol",
    "maxDD",
    "annExcessReturn",
    "TE",
    "IR",
]

rename_columns = {
    "alpha": "Alpha",
    "beta": "Beta",
    "ariMeanRt": "Arithmetic Mean Return",
    "geoMeanRt": "Geometric Mean Return",
    "cagr": "CAGR",
    "sharpe": "Sharpe Ratio",
    "sortino": "Sortino Ratio",
    "annVol": "Annualized Volatility",
    "maxDD": "Max Drawdown",
    "annExcessReturn": "Annualized Excess Return",
    "TE": "Tracking Error",
    "IR": "Information Ratio",
}


fig, axs = plt.subplots(4, 3, figsize=(8, 9))

for column_of_interest, ax in zip(columns_of_interest, axs.flatten()):
    data_original = data_to_plot_original[
        (
            data_to_plot_original[column_of_interest]
            < data_to_plot_original[column_of_interest].quantile(0.975)
        )
        & (
            data_to_plot_original[column_of_interest]
            > data_to_plot_original[column_of_interest].quantile(0.025)
        )
    ]
    data_cloned = data_to_plot_cloned[
        (
            data_to_plot_cloned[column_of_interest]
            < data_to_plot_cloned[column_of_interest].quantile(0.975)
        )
        & (
            data_to_plot_cloned[column_of_interest]
            > data_to_plot_cloned[column_of_interest].quantile(0.025)
        )
    ]

    ciks_original = data_original["cik"].to_list()
    ciks_cloned = data_cloned["cik"].to_list()
    ciks_intersection = list(set(ciks_original) & set(ciks_cloned))

    data_original = data_original[data_original["cik"].isin(ciks_intersection)][
        column_of_interest
    ]
    data_cloned = data_cloned[data_cloned["cik"].isin(ciks_intersection)][
        column_of_interest
    ]

    # set both axes to the same limits: -100, 320
    x_min = min(data_original.min(), data_cloned.min())
    x_max = max(data_original.max(), data_cloned.max())
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(x_min, x_max)

    ax.scatter(
        data_original,
        data_cloned,
        s=1,
    )

    # plot line of equality
    x = np.linspace(x_min, x_max)
    ax.plot(x, 1 * x + 0, color="red", linestyle="--", linewidth=0.8)

    ax.set_xlabel("Original Funds")
    ax.set_ylabel("Cloned Funds")
    ax.set_title(f"{rename_columns[column_of_interest]}")
    ax.yaxis.grid(True, linestyle="--", which="major", color="grey", alpha=0.25)
    ax.xaxis.grid(True, linestyle="--", which="major", color="grey", alpha=0.25)
    ax.axhline(y=0, color="black", alpha=0.5, linestyle="--", linewidth=0.8)
    ax.axvline(x=0, color="black", alpha=0.5, linestyle="--", linewidth=0.8)


plt.tight_layout()
file_path = "./assets/factors-original-cloned.png.pdf"
plt.savefig(file_path, dpi=300, format="pdf")
print(f"Saved to {file_path}")

plt.show()
```
